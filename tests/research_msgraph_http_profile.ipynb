```python
sql = (
    '''
    SELECT
        [Id],
		[GUID],
		CAST([CreatedDatetime] AS NVARCHAR(MAX)) AS CreatedDatetime,
        CAST([ModifiedDatetime] AS NVARCHAR(MAX)) AS ModifiedDatetime,
		[ClientId],
		[Tenant],
		[ClientSecret],
		[AccessToken],
		[ExpiresIn],
		[ExtExpiresIn],
		[RefreshToken],
		[Scope],
		[TokenType],
		[UserId]
	FROM ms.Auth;
    '''
)
sql
```

```python
row = cursor.execute(sql).fetchall()
row
```

```python
# Get column names from the cursor description
column_names = [desc[0] for desc in cursor.description]

# Loop through each tuple in row and print attribute name and value
for record in row:
    for name, value in zip(column_names, record):
        print(f"{name}: {value}")
    print("-" * 40)
```

```python
from azure.identity import ClientSecretCredential
```

```python
from msgraph import GraphServiceClient
```

```python
my_auth_record = row[0]
my_auth_record
```

```python
client_id = my_auth_record.__getattribute__('ClientId')
client_id
```

```python
tenant_id = my_auth_record.__getattribute__('Tenant')
tenant_id
```

```python
client_secret = my_auth_record.__getattribute__('ClientSecret')
client_secret
```

```python
access_token = my_auth_record.__getattribute__('AccessToken')
access_token
```

```python
credential = ClientSecretCredential(
    tenant_id=tenant_id,
    client_id=client_id,
    client_secret=client_secret
)
credential
```

```python
client = GraphServiceClient(credentials=credential)
client
```
