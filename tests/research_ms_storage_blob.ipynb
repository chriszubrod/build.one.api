```python
# Local file to upload

local_file_path = "/Users/chris/Downloads/COIs - 02/A&A Masonry - GL - 2024.12.14 - 2025.12.14"
local_file_path
```

```python
# Define blob path (your convention)
doc_id = "3f8d3e59-27a6-41ba-b1c7-82f1d0e29f1e"
blob_name = f"vendor/certificate/gl/{doc_id}.pdf"
blob_name
```

```python
# Define metadata (stored as key-value, visible in portal)
metadata = {
    "doctype": "certificate",
    "category": "gl",
    "docid": doc_id,
    "vendor": "AcmeCorp"
}
type(metadata)
```

```python
# Define blob index tags (queryable at storage layer, case-sensitive keys)
tags = {
    "doctype": "certificate",
    "category": "gl",
    "docId": doc_id,
    "vendor": "AcmeCorp"
}
type(tags)
```

```python
# Optional: set proper content type
content_settings = ContentSettings(content_type="application/pdf")
```

```python
# Upload with metadata
with open(local_file_path, "rb") as data:
    container_client.upload_blob(
        name=blob_name,
        data=data,
        overwrite=True,
        content_settings=content_settings,
        metadata=metadata
    )
```

```python
# Apply tags (must be a separate call)
blob_client = container_client.get_blob_client(blob_name)
blob_client.set_blob_tags(tags)
```

```python
print(f"Uploaded {local_file_path} as {blob_name} with metadata + tags.")
```

```python
for tagged_blob in container_client.find_blobs_by_tags("doctype = 'certificate' AND category = 'gl'"):
    print(tagged_blob['name'])
```
