```python
sql = (
    '''
    SELECT
        [Id],
		[GUID],
		CAST([CreatedDatetime] AS NVARCHAR(MAX)) AS CreatedDatetime,
        CAST([ModifiedDatetime] AS NVARCHAR(MAX)) AS ModifiedDatetime,
		[ClientId],
		[Tenant],
		[ClientSecret],
		[AccessToken],
		[ExpiresIn],
		[ExtExpiresIn],
		[RefreshToken],
		[Scope],
		[TokenType],
		[UserId]
	FROM ms.Auth;
    '''
)
sql
```

```python
row = cursor.execute(sql).fetchall()
row
```

```python
# Get column names from the cursor description
column_names = [desc[0] for desc in cursor.description]

# Loop through each tuple in row and print attribute name and value
for record in row:
    for name, value in zip(column_names, record):
        print(f"{name}: {value}")
    print("-" * 40)
```
